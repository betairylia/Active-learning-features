```python
from opacus.grad_sample import GradSampleModule
import torch
from torch import Tensor
from torch import nn
```

```python
net = torch.nn.Sequential(
    torch.nn.Flatten(),
    torch.nn.Linear(4, 15),
    torch.nn.ReLU()
)

head = torch.nn.Sequential(
    torch.nn.Linear(15, 4)
)
```

```python
xs = torch.randn(size = (12, 4))
```

```python
def patch_module(seq):
    for i in range(len(seq)):
        if isinstance(seq[i], nn.Linear):
            seq[i] = GradSampleModule(seq[i])

def unpatch_module(seq):
    for i in range(len(seq)):
        if isinstance(seq[i], GradSampleModule):
            seq[i] = next(seq[i].children())
            
patch_module(net)
patch_module(head)
```

```python
print(xs)

net.zero_grad()
head.zero_grad()

f = net(xs)
l = head(f)
loss = l.sum()
```

```python
loss.backward()
```
